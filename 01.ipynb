### Put all npy files together in single folder and keep path in `base_path` variable 


```python
import os
import numpy as np
import mne

ch_names = ['Fp1', 'Fp2', 'F3', 'F4', 'C3', 'C4', 'P3', 'P4', 'O1', 'O2', 'F7', 'F8', 'T3', 'T4', 'T5','T6','Fz', 'Cz', 'Pz','PG1', 'PG2', 'A1', 'A2','ECG1', 'ECG2','EMG1','EMG2', 'EMG3', 'EMG4']
ch_types = ["eeg"] * len(ch_names)


base_path = r"E:\DATASETS\opensource-dataset\NPY-Files"
output_path = r"E:\DATASETS\opensource-dataset\EDF-Files"
files = os.listdir(base_path)
def split(fl:str)->list:
    parts = fl.split("_")
    return [parts[0] , int(parts[1]) , int(parts[2])]
    
    
files= sorted(files, key=split)
print(len(files))
output_files= {}
for file in files:
    f= file.strip().split("_")[0]
    if f not in output_files:
        output_files[f] = list(filter(lambda fl: str(fl).startswith(f), files))
    # output_files[f] = list(filter(lambda fl: str(fl).startswith(f), files))
```

```python
for fl_name , npy_files in  output_files.items():
    A = None
    for npy_file in npy_files:
        if A is None:
            A = np.load(os.path.join(base_path , npy_file))
        else:
            A = np.concatenate((A,  np.load(os.path.join(base_path , npy_file))) , axis = 1)

    sfreq = 500 # Hz
    ch_names = ['Fp1', 'Fp2', 'F3', 'F4', 'C3', 'C4', 'P3', 'P4', 'O1', 'O2', 'F7', 'F8', 'T3', 'T4', 'T5','T6','Fz', 'Cz', 'Pz','PG1', 'PG2', 'A1', 'A2','ECG1', 'ECG2','EMG1','EMG2', 'EMG3', 'EMG4']
    ch_types = ["eeg"] * len(ch_names)
    info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types=ch_types)
    raw = mne.io.RawArray(A, info)
    raw.export(os.path.join(output_path,  fl_name +  ".e"), fmt="edf", overwrite=True)
```
